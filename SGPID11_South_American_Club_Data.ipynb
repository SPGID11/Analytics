```python
#@title Imports
from plotnine import *
import numpy as np
import pandas as pd
import re

import warnings
warnings.filterwarnings('ignore')


pd.set_option("display.max_columns", 999)
```

```python
#downloads dictionaries into files
uruguay = pd.read_csv("uruguay_clubs.csv")
peru = pd.read_csv("peru_clubs.csv")
chile = pd.read_csv("chile_clubs.csv")
clubs = uruguay.append(peru, ignore_index=True)
clubs = clubs.append(chile, ignore_index=True)
clubs = pd.DataFrame(clubs)
```

```python
#@title Cleaning

#converting string variables to floats if needed
def remove_units_and_convert_to_float(value):
    if isinstance(value, str):
      numeric_part = re.sub(r'[^-0-9.]', '', value)
      return float(numeric_part)
    elif isinstance(value, int):
      return value
    else:
      return float(value)

variables_to_convert = ['squad', 'avgAge', 'foreigners', 'avgMarketValue', 'totalMarketValue', 'totalMarketValue_2022', 'totalMarketValue_2023', 'followers', 'trophies','avgAttendance', 'capacity_k']
clubs[variables_to_convert] = clubs[variables_to_convert].applymap(remove_units_and_convert_to_float)

#remove unnecessary columns
columns_to_remove = ['url', 'id']
clubs = clubs.drop(columns=columns_to_remove)

#renaming columns
new_column_names = {
    'avgMarketValue': 'avgMarketValue_k',
    'totalMarketValue_2022': 'marketValue_2022_m',
    'totalMarketValue_2023': 'marketValue_2023_m',
    'totalMarketValue': 'marketValue_m',
    'followers': 'followers_k'
}

clubs = clubs.rename(columns = new_column_names)
```
